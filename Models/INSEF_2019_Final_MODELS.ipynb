```python
# -*- coding: utf-8 -*-
"""
Created on Mon Dec 17 17:37:02 2018

@author: tarang,devansh
"""
#importing libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from keras.models import Sequential
from keras.layers import Dense
from joblib import dump, load
from copy import deepcopy
from collections import Counter 
import random

#setting up a seed to as to provide the same distribution every time
random.seed = 9999

#reading the data
data = pd.read_csv('heart.csv')

x_data = data.drop(columns = ['target'],axis=1)
y = data.target.values

# standard scaling
x = (x_data - np.mean(x_data))/np.std(x_data)

#dividing the data in train and test
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)

def Logistic_Regression():
    
    global x_train,x_test,y_train,y_test
    
    model = LogisticRegression()
    model.fit(x_train,y_train)
    
    #returning object & accuracy
    return {'model':model,
            'accuracy':model.score(x_test,y_test)*100}
    

def KNN():
    
    global x_train,x_test,y_train,y_test
    
    model = KNeighborsClassifier(n_neighbors = 24)
    model.fit(x_train,y_train)
        
    #returning object & accuracy
    return {'model':model,
            'accuracy':(model.score(x_test,y_test)*100)}
    
    

def DecisionTree():
    
    global x_train,x_test,y_train,y_test
    
    model = DecisionTreeClassifier(random_state=0)
    model.fit(x_train,y_train)
    
    #returning object & accuracy
    return {'model':model,
            'accuracy':model.score(x_test,y_test)*100}
    
    

def SVM_Classifier():
    
    
    global x_train,x_test,y_train,y_test
    
    model = svm.LinearSVC()
    model.fit(x_train,y_train)
    
    #returning object & accuracy
    return {'model':model,
            'accuracy':model.score(x_test,y_test)*100}
    

def ANN():
    
    
    global x_train,x_test,y_train,y_test
    
    model = Sequential()
    
    #implicit input layer combined with hidden layer
    model.add(Dense(units = 13, kernel_initializer = 'uniform', activation = 'relu', input_dim = 13))
    
    #hidden layer 2
    model.add(Dense(units = 20, kernel_initializer = 'uniform', activation = 'relu', input_dim = 13))
    
    #output layer
    model.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    
    model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    
    #fitting with optimal hyperparameters
    model.fit(x_train, y_train, batch_size = 25, nb_epoch = 300)
    
    return {'model':model,
            'accuracy':accuracy_score(model.predict(x_test) > 0.5, y_test)*100}
    
    

def NaiveBayes():
    
    global x_train,x_test,y_train,y_test
    
    model = BernoulliNB()
    
    model.fit(x_train,y_train)
    
    return {'model':model,
            'accuracy':accuracy_score(model.predict(x_test) > 0.5, y_test)*100}
    

'''------------------------------------------------------------------------------------------------------'''

lr = Logistic_Regression()
dt = DecisionTree()
knn = KNN()
ann = ANN()
nb = NaiveBayes()
svm = SVM_Classifier()

#printing
print('lr:',lr['accuracy'],
      '\ndt:',dt['accuracy'],
      '\nknn:',knn['accuracy'],
      '\nann:',ann['accuracy'],
      '\nnb:',nb['accuracy'],
      '\nsvm:',svm['accuracy'])


    
```

```python
#saving models
dump(lr,"lrFinal.joblib")
dump(dt,"dtFinal.joblib")
dump(knn,"knnFinal.joblib")
dump(ann,"annFinal.joblib")
dump(nb,"nbFinal.joblib")
dump(svm,"svmFinal.joblib")
```
